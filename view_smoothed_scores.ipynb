### View all of the smoothed college confessions sentiment scores 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
caltech_smoothed = "caltech_smoothed.csv"
df = pd.read_csv(caltech_smoothed, header=0, sep=u",", index_col = None, encoding="ISO-8859-1")
df.head()
```

```python
def plot_metric(score_name, df):
    """
    Helper function to plot a smoothed metric over time 
    """
    timestamps = df["timestamp"].tolist()
    scores = df[score_name].tolist()
    plt.plot(timestamps, scores)
    plt.title("Sentiment " + score_name + " score vs. Timestamp")
    plt.show()
```

```python
plot_metric("compound_smoothed", df)
```

```python
plot_metric("positive_smoothed", df)
```

```python
plot_metric("negative_smoothed", df)
```

```python
plot_metric("neutral_smoothed", df)
```
